IMPORTING IMPORTANT LIBRARIES:

```python
import pandas as pd  #IMPORTING DATASET
from sklearn.tree import DecisionTreeClassifier   #FOR DECISION TREE CLASSIFIER
from sklearn.model_selection import train_test_split  #FOR SPLITING TRAINING AND TESTING DATA
from sklearn import datasets
from sklearn import tree     #VISUALIZE TREE
import matplotlib.pyplot as plt    #FOR VISUALIZATION
```

LOAD THE DATASET:

```python
data = pd.read_csv("Iriss.csv")    #LOADS DATASET
data
```

```python
data.head()   #FIRST FEW ROWS OF DATASET
```

```python
data.tail()   #LAST FEW ROWS OF DATASET
```

```python
data.info()    #INFORMATION ABOUT THE DATASET
```

```python
data.describe()       #DISPLAYS STATISTICAL VALUES
```

SPLIT DATA INTO FEATURE(X) AND TARGET VARIABE(Y):

```python
x = data.drop("Species", axis=1)     #FEATURES
y = data["Species"]                  #TARGET VARIABLE
```

SPLIT THE DATASET INTO TRAINING AND TESTING DATA:

```python
# 70% DATA TO ('x_train','y_train')  & 30% DATA TO ('x_test','y_test')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

EVALUATE THE CLASSIFIER:

```python
classifier = DecisionTreeClassifier()   #CREATW AN INSTANCE OF DECISION_TREE_CLASSIFIER CALLED AS CLASSIFIER
```

TRAIN THE CLASSIFIER:

```python
classifier.fit(x_train, y_train)  #TRAINS THE CLASSIFIER ON TRAINING DATA
```

EVALUATE THE CLASSIFIER:

```python
accuracy = classifier.score(x_test, y_test) #TEST THE CKASSIFIER ON TESTING DATA
print("Accuracy:", accuracy)                #ACCURACY SCORE IS PRINTED
```

VISUALIZE THE TREE:

```python
fig, ax = plt.subplots(figsize=(12,22))
tree.plot_tree(classifier, ax=ax)
plt.show
```

```python
plt.figure(figsize=(14,9))
tree.plot_tree(classifier, filled=True, feature_names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
```

ADDING & DEFINING NEW DATA:

> This cell defines new data points (new_data) to be used for prediction.

>  It consists of three data points, each containing values for the features used in the decision tree.












```python
new_data = [[5.1, 3.5, 1.4, 0.2, 2.2], [6.2, 3.2, 2.8, 4.8, 1.8], [7.3, 7.6, 3.2, 6.5, 2.3]]
```

PREDICT THE CLASS FOR NEW DATA:

> Predicts the class for the new data points (new_data) using the trained decision tree classifier (classifier.predict()).

>  The predicted classes are stored in the predicted_classes variable.






```python
predicted_classes = classifier.predict(new_data)
```

PRINT THE PREDICTED CLASSES:
> Prints the predicted classes for each data point in new_data

```python
for i, predicted_class in enumerate(predicted_classes):
    print(f"Data point {i+1}: Predicted class is {predicted_class}")
```
